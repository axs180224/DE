```python
import numpy as np 
import pandas as pd
file_path = "C:\\Users\\anand\\Downloads\\imdb_top_1000.csv"
df = pd.read_csv(file_path)
```

```python
#Get rid of uselsss columns
new_df = df.drop(['Poster_Link', 'Certificate', 'Overview', 'Meta_score', 'No_of_Votes'], axis =1)
new_df.head()
```

```python
# Oldest movie
print(new_df['Released_Year'].min())
df_oldest = new_df.sort_values('Released_Year')
df_oldest.iloc[0]
```

```python
# Newest movie
df_oldest.iloc[998]
```

```python
# Top 10 movies by IMDB rating
IMDB_df = new_df.sort_values('IMDB_Rating', ascending = False)
IMDB_df.head(10)
```

```python
# Top movie for each genre
movies = new_df['Genre'].str.split(', ')
```

```python
# Find unique genres
unique_genres = set()
for movie in movies:
    unique_genres.update(movie)

u_genres = list(unique_genres)
u_genres
```

```python
#print out each genres highest rated movie
for genre in u_genres:
    best_movie = df.loc[new_df['Genre'].str.contains(genre)].nlargest(1,'IMDB_Rating')
    print(f"Highest rated movie for {genre} is {best_movie['Series_Title'].values[0]}")
```

```python
#Director with the most movies
new_df['Director'].value_counts().head(1)
```

```python
#star with the most movies
new_df['Star1'].value_counts().head(1)
```

```python
new_df = new_df.dropna()
new_df.info()
```

```python
new_df['Gross'] = new_df['Gross'].str.replace(',', '')
new_df['Gross'] = new_df['Gross'].astype(int)
new_df.dtypes
```

```python
#Highest grossing movie for each genre 
for genre in u_genres:
    highest_gross = new_df.loc[new_df["Genre"].str.contains(genre)]
    highest_gross = highest_gross.nlargest(1, "Gross")
    print(f" {genre} : {highest_gross['Series_Title'].values[0]}")
```

```python
#Lowest grossing for each director
u_directors = new_df["Director"].unique()

for director in u_directors:
    lowest = new_df.loc[new_df["Director"].str.contains(director)]
    lowest = lowest.nsmallest(1, "Gross")
    print(f"{director} : {lowest['Series_Title'].values[0]}")
```
