```python
# Feature Engineering

df['Normal Stats'] = df['Total']/df['Total'].max()
df.head()
```

```python
# Alter multiple values at once
df.loc[:3, 'Type 1'] = 'Fire'
df.head()
```

```python
#Sorting our DF .sort_values()

df.sort_values('Total', inplace = True, ascending= False)
df.head()
```

```python
df.reset_index(inplace = True, drop  = True)
df.head()
```

```python
# Combining Conditions

df = df[(df['Type 1'] == 'Fire') & (df['Type 1'] == 'Flying')].head(15)
df.head()
```

# Creating a Pivot Table
- Used multi-variate analysis
- Allows us to see metrics across categorical data
- Pivot tables need to consider Cardinality 
    - Cardinality = number of possible values

```python
# use nuniques to determine cardinality

df.nunique()
```

```python
pivot_table = df.pivot_table(
values=['Total', 'HP', 'Attack','Defense', 'Sp. Atk'],
index = ['Type 1', 'Type 2']
)
#pivot_table.head()
pivot_table.sort_values(['Total'], ascending=False).head()
#Pivoted type 1 to be a row
```
